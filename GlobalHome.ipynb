```python
from urllib.request import urlopen

androidurl3 = 'https://digitalmarketingdeal.com/blog/construction-companies-in-chennai/'
android_data3 = urlopen(androidurl3)
##android_data3

## Getting the whole html page
android_html3 = android_data3.read()
##print(android_html3)
android_data3.close()

from bs4 import BeautifulSoup as  soup

android_soup3 = soup(android_html3,'html.parser')
android_soup3

name3 = [words.text[2:] for words in android_soup3.findAll('b',{})]
name3 = name3[1:]
##name3

address_i = [words.text for words in android_soup3.findAll('span',{})]
##address_i

import re

address3 = []
for addr in address_i:
    if(re.search('^Addr',addr)):
        address3.append(addr)
##address3

for word in address3:
    if word == 'Address:':
        address3.remove(word)
##address3

import re

contacts3 = []
for conts in address_i:
    if(re.search('^Phone',conts)):
        contacts3.append(conts)
##contacts3

import re

websites3 = []
for webs in address_i:
    if(re.search('^Webs',webs)):
        websites3.append(webs)
##websites3

for word in websites3:
    if word == 'Website:':
        websites3.remove(word)
##websites3
```
